### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
import pandas as pd 
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
df.head()
```

```python
# Dividing data into X and y

X = df.iloc[:,:].values
y = df.iloc[:,4].values
print(X.shape)
print(y.shape)
```

```python
label_encoder_y = LabelEncoder()
y = label_encoder_y.fit_transform(y)
```

```python
#Standard Scaling

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
#Tranforming data into 3Dimesional
pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

```python
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y,cmap=plt.cm.tab10_r)
ax.set_title("First three PCA directions")

plt.show()
```
