# We have the min and max temperatures in a city In India for each months of the year.  
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize as os
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
months = np.arange(12)
plt.plot(months, max_temp, 'bs')
plt.plot(months, min_temp, 'rs')
plt.xlabel('Months')
plt.ylabel('Min(blue) and max(red) temperature')
```

```python
def yearly_temps(t, avg, ampl, t_offset):
    return (avg
            + ampl * np.cos((t + t_offset) * 2 * np.pi / t.max()))
res_max, cov_max = os.curve_fit(yearly_temps, months, max_temp, [20, 10, 0])
res_min, cov_min = os.curve_fit(yearly_temps, months, min_temp, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)
plt.figure(figsize=(8,6))
plt.plot(months, max_temp, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature')

plt.show()
```

## 1. Create a pie chart presenting the male/female proportion                                                     2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
gender = titanic['sex'].astype('category')
gender = gender.cat.codes
gender.head()
```

```python
mycolors = ["Cyan", "Yellow"]
plt.pie(gender.value_counts(), labels = titanic['sex'].unique(), autopct='%1.1f%%',colors = mycolors)
plt.title('Male/Female Proportion')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(titanic['age'], titanic['fare'], c = gender)
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.show()
```
