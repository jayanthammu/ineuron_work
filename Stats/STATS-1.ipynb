# Problem Statement 3:
## The number of times I go to the gym in weekdays, are given below along with its associated probability:
## x = 0, 1, 2, 3, 4, 5
## f(x) = 0.09, 0.15, 0.40, 0.25, 0.10, 0.01
## Calculate the mean no. of workouts in a week. Also evaluate the variance involved in it.

```python
x= np.asarray([0, 1, 2, 3, 4, 5])
fx = np.array( [0.09, 0.15, 0.40, 0.25, 0.10, 0.01])
```

```python
mean=np.dot(x,fx)
variance_of_x=(x-mean)**2
variance = np.dot(variance_of_x.reshape(1,-1),fx)
print('mean no. of workouts in a week:', mean)
print('Variance of workouts in a week:', variance)
```

# Problem Statement 5:

### A company manufactures LED bulbs with a faulty rate of 30%. If I randomly select 6 chosen LEDs, what is the probability of having 2 faulty LEDs in my sample? Calculate the average value of this process. Also evaluate the standard deviation associated with it.

```python
import scipy.special
x = 0.3 #Faulty
y = 0.7  #Not Faulty
df=pd.DataFrame({'a':[int(i) for i in range(7)],
                 'B_a':[scipy.special.comb(6,i)*(x**i)*(y**(6-i)) for i in range(7)]})
print(df.iloc[2])
plt.figure(figsize=(15,5))
sns.barplot('a','B_a',data=df,color = 'blue')
plt.xlabel('Number of Faulty Leds')
plt.ylabel('Probability of Fault')
```

```python
df['Expected value']=df['a']*df['B_a']
mean = np.round(df['Expected value'].sum())
print('mean =',mean)
df['variance'] = df['B_a']*(df['a']-mean)**2
std = np.sqrt(df['variance'].sum())
print('Standard Deviation :',np.round(std))
```

# Problem Statement 7:

Customers arrive at a rate of 72 per hour to my shop. What is the probability of k customers arriving in 4 minutes? a) 5 customers, b) not more than 3 customers, c) more than 3 customers. Give a pictorial representation of the same to validate your answer.

```python
from scipy.stats import poisson
mu = 4*(72/60)
print('The probability of arriving 5 cutomers in 4 minutes is :', poisson.pmf(k=5,mu=mu))
print('The probability of arriving not more than 3 customers in 4 minutes is : ',poisson.pmf(k=3, mu=mu))
print('The Probability of more than 3 customers arriving in 4 minutes is : ',-poisson.cdf(k=3,mu=mu))
```

```python
d_bin = poisson.rvs(mu=4.8, size=200)
ax = sns.distplot(d_bin)
ax.set(xlabel='Customers per 4 minutes', ylabel='Frequency')
plt.title('Poisson distribution')
plt.show()
```
